funzioni per estrazione delle colonne json

```python
#funzione per l'estrazione del file json
def extract_json(df, column):
    result = df[column].apply(pd.read_json) 
    movie_id = df['movie_id'] 
    for i in range(len(df)):
        result[i]['movie_id'] = movie_id[i] 
    return pd.concat(result.values).reset_index().drop("index", axis = 1)
```

```python
#funzione per l'estrazione del file json
def extract_json1(df, column):
    result = df[column].apply(pd.read_json) 
    movie_id = df['id'] 
    for i in range(len(df)):
        result[i]['movie_id'] = movie_id[i] 
    return pd.concat(result.values).reset_index().drop("index", axis = 1)
```

4) Extract the domain of each homepage

```python
nh = movies[movies['homepage'].notnull()].reset_index().drop('index', axis = 1)
def extract_domain(a):
    domain = [] #[]perchè è una lista, se fosse dizionario metterei {}
    homepage = a.split() #definisco homepage
    for i in range(len(homepage)):
        domain.append(homepage[i].split('/')[2].split('.')[-1]) #list.append per aggiungere qualcosa alla lista e splitto le parti che non mi interessano 
    return str(domain)[2:-2] #seleziono la parte della stringa che voglio da ['com']
nh['domain'] = nh['homepage'].apply(extract_domain)
nh[['homepage', 'domain']].head()
```

13) Distribute the revenue according to the order of appearance in a movie. Assume that the i-th actor contributes twice as much as the (i+1)-th actor to the revenue

```python
def assign_part(orders,number_of_actors,revenues):
   
    part = [0]*len(orders)
    for i in range(len(orders)):
        k = orders[i]
        n = number_of_actors[i]
        rev = revenues[i]
        e_n = np.exp2(n)
        e_k = np.exp2(k)
        
        number = rev/e_k + rev/(e_k * (e_n -1))
        part[i] = np.round(number/2)
        
    return part
pd.set_option('display.float_format', lambda x: '%.2f' % x)

uniti4['part'] = assign_part(uniti4['order'],uniti4['number_of_actors'],uniti4['revenue'])
uniti4.head()
```

14) For each actor find the total revenue attributed to him/her

```python
revenue_each_actors = uniti4.groupby(['actor_id', 'name']).sum()['part']
revenue_each_actors1 = revenue_each_actors.to_frame().reset_index().rename(columns = {'part' : 'total_revenue'})
revenue_each_actors1.head()
```

15) Find the actor that is responsible for the most overall revenue

```python
top_actor = revenue_each_actors1.loc[revenue_each_actors1['total_revenue'].idxmax()]
top_actor
```
